```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTENC
```

```python
credito = pd.read_csv('credit_simple.csv', sep=';')
credito.head()
```

```python
count = credito.groupby('CLASSE').size()
count
```

```python
y = credito['CLASSE'].values
x = credito.iloc[:,:-1].values
```

```python
labelencoder = LabelEncoder()
for i in range(x.shape[1]):
    if x[:,i].dtype == 'object':
        x[:,i] = labelencoder.fit_transform(x[:,i])
```

```python
sm = SMOTENC(categorical_features=[3,5,6], random_state=42)

x_res, y_res = sm.fit_resample(x, y)
```

```python
x_res
```

```python
df = pd.DataFrame({'CLASSE': y_res})
df.value_counts()
```
